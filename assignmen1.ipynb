```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel('carbonDb.xlsx', sheet_name='PROJECTS', header=[1], engine='openpyxl')
```

```python
df.describe()
```

```python
for i in df.columns:
    print(i)
```

```python
c1 = 'Credits issued by vintage year (when reduction/removals occurred)'
c2 = 'Credits retired or cancelled in:'
credits_by_vintage = df.loc[:, c1:c2].iloc[:, :-1]
credits_by_vintage.columns = credits_by_vintage.iloc[0]
credits_by_vintage = credits_by_vintage[1:]
credits_by_vintage
```

```python
df.columns = df.iloc[0]
df = df[1:]
df.columns = df.columns.astype(str)
df.rename(columns=lambda x: x.replace('\n', ''), inplace=True)
for i in df.columns:
    print(i)
```

```python
# df = df[['Project ID', 'Project Name', 'Voluntary Registry' 'Scope', 'Type','Reduction / Removal', 'Region','Country', 'Total Credits Issued', 'Total Credits Retired','Total Credits Remaining', 'First Year of Project (Vintage)', 'Project Type From the Registry',]]

df.rename(columns={'First Year of Project (Vintage)': 'Vintage', ' Scope': 'Scope', ' Type': 'Type', 'Project Type From the Registry': 'Project Type'}, inplace=True)
df = df[['Project ID', 'Project Name', 'Voluntary Registry', 'Scope', 'Type', 'Reduction / Removal', 'Region', 'Country', 'Total Credits Issued', 'Total Credits Retired', 'Total Credits Remaining', 'Vintage', 'Project Type', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']]
df = df.dropna(subset=['Vintage'])
df = df.loc[:, ~df.columns.duplicated()]
df
```

```python
df.describe()
```

```python
total_credits = df.groupby(['Project Type', 'Country'])['Total Credits Issued'].sum()

# Analysis

total_credits_by_country_and_type = df.groupby(['Project Type', 'Country'])['Total Credits Issued'].sum()
print("Total Credits Issued by Project Type and Country:")
total_credits_by_country_and_type
```

```python
project_trends = df.pivot_table(index='Project Type', values=df.columns[13:], aggfunc='sum')
project_trends
```

```python
df['Total Credits Issued'] = pd.to_numeric(df['Total Credits Issued'], errors='coerce')
projects_highest_credit = df.nlargest(5, 'Total Credits Issued')
projects_highest_credit
```

```python
projects_highest_credit = df.nsmallest(5, 'Total Credits Issued')
projects_highest_credit
```
